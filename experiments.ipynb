```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

iris = pd.read_csv("./iris_csv.csv")

X = iris[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']]
y = iris["class"]

plt.hist(y)
plt.show()
```

```python
colors = y.map({
    "Iris-virginica": "red",
    "Iris-versicolor": "blue",
    "Iris-setosa": "yellow",
})

plt.scatter(X['sepallength'], X['sepalwidth'], marker='o', color=colors)
plt.show()
```

```python
colors = y.map({
    "Iris-virginica": "red",
    "Iris-versicolor": "blue",
    "Iris-setosa": "yellow",
})

plt.scatter(X['petallength'], X['petalwidth'], marker='o', color=colors, label=y)
plt.legend()
plt.show()
```

```python
from cnbc.cnbc import CNBC, PointClass


c = CNBC(k=8)

must_link = []
cannot_link = []

cluster = c.fit(X, must_link, cannot_link)

plt.hist(cluster.astype(str))
plt.show()
```

```python
import pylab
import numpy as np

colors = cluster.replace({
    PointClass.NOISE: -1
})

fig = plt.figure()
plt.scatter(X['sepallength'], X['sepalwidth'], c=colors, label=colors)
plt.legend()
plt.show()

```
