```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

iris = pd.read_csv("./iris_csv.csv")

X = iris[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']]
y = iris["class"]

plt.hist(y);
plt.show()
```

```python
colors = y.map({
    "Iris-virginica": "red",
    "Iris-versicolor": "blue",
    "Iris-setosa": "yellow",
})

plt.scatter(X['sepallength'], X['sepalwidth'], marker='o', color=colors)
plt.show()
```

```python
colors = y.map({
    "Iris-virginica": "red",
    "Iris-versicolor": "blue",
    "Iris-setosa": "yellow",
})

plt.scatter(X['petallength'], X['petalwidth'], marker='o', color=colors)
plt.show()
```
