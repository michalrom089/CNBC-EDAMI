```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

iris = pd.read_csv("./iris_csv.csv")

X = iris[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']]
y = iris["class"]

plt.hist(y)
plt.show()
```

```python
y_c = y.map({
    "Iris-virginica": 0,
    "Iris-versicolor": 1,
    "Iris-setosa": 2,
})

sc = plt.scatter(X['petallength'], X['petalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica",  "Iris-versicolor", "Iris-setosa"])
plt.show()
```

```python
y_c = y.map({
    "Iris-virginica": 0,
    "Iris-versicolor": 1,
    "Iris-setosa": 2,
})

sc = plt.scatter(X['sepallength'], X['sepalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica", "Iris-versicolor", "Iris-setosa"])
plt.show()
```

```python
from cnbc.cnbc import CNBC, PointClass
from cnbc.view import plot_cnbc_clusters
import pylab
import numpy as np

c = CNBC(k=8)

must_link = []
cannot_link = []

cluster = c.fit(X, must_link, cannot_link)

plt.hist(cluster.astype(str))
plt.show()
```

```python
c = CNBC(k=8)

must_link = []
cannot_link = []

cluster = c.fit(X, must_link, cannot_link)

x = X['sepallength']
y = X['sepalwidth']
classes = cluster

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plot_cnbc_clusters(x, y, cluster)

plt.subplot(1,2,2)
plt.scatter(x, y, marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica", "Iris-versicolor", "Iris-setosa"])

plt.show()

```
