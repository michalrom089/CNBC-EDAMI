```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

iris = pd.read_csv("./iris_csv.csv")

X = iris[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']]
y = iris["class"]
y_c = y.map({
    "Iris-virginica": 0,
    "Iris-versicolor": 1,
    "Iris-setosa": 2,
})

plt.hist(y)
plt.show()
```

```python
sc = plt.scatter(X['petallength'], X['petalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica",  "Iris-versicolor", "Iris-setosa"])
plt.show()
```

```python
sc = plt.scatter(X['sepallength'], X['sepalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica", "Iris-versicolor", "Iris-setosa"])
plt.show()
```

```python
from cnbc.cnbc import CNBC, PointClass
from cnbc.view import plot_cnbc_clusters
import pylab
import numpy as np

c = CNBC(k=8)

must_link = []
cannot_link = []

cluster = c.fit(X, must_link, cannot_link)

plt.hist(cluster.astype(str))
plt.show()
```

```python
c = CNBC(k=8)

must_link = []
cannot_link = []

cluster = c.fit(X, must_link, cannot_link)


plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.title("Sepal CNBC")
plot_cnbc_clusters(X['sepallength'], X['sepalwidth'], cluster)

plt.subplot(2,2,2)
plt.title("Sepal Test")
plt.scatter(X['sepallength'], X['sepalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica", "Iris-versicolor", "Iris-setosa"])

plt.subplot(2,2,3)
plt.title("Petal CNBC")
plot_cnbc_clusters(X['petallength'], X['petalwidth'], cluster)

plt.subplot(2,2,4)
plt.title("Petal Test")
plt.scatter(X['petallength'], X['petalwidth'], marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica", "Iris-versicolor", "Iris-setosa"])

plt.show()

```

```python
x = X['petallength']
y = X['petalwidth']

plt.figure(figsize=(16,16))
sc = plt.scatter(x, y, marker='o', c=y_c)
plt.legend(handles=sc.legend_elements()[0], labels=["Iris-virginica",  "Iris-versicolor", "Iris-setosa"])
for i in range(len(x)):
    plt.text(x[i], y[i], '({})'.format(i))
plt.show()
```

```python
c = CNBC(k=12)

must_link = [(22,13), (22,12)]
cannot_link = [(43,23)]

constraints_cluster = c.fit(X, must_link, cannot_link)
cluster = c.fit(X, [], [])

plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
plt.title("Sepal CNBC")
plot_cnbc_clusters(X['sepallength'], X['sepalwidth'], cluster)

plt.subplot(2,2,2)
plt.title("Sepal CNBC - constraints")
plot_cnbc_clusters(X['sepallength'], X['sepalwidth'], constraints_cluster)

plt.subplot(2,2,3)
plt.title("Petal CNBC")
plot_cnbc_clusters(X['petallength'], X['petalwidth'], cluster)

plt.subplot(2,2,4)
plt.title("Petal CNBC - constraints")
plot_cnbc_clusters(X['petallength'], X['petalwidth'], constraints_cluster)

plt.show()

```
